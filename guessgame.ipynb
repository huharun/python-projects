```python
import random

def guessingGame(guessLimit, number):
    random_number = random.randint(1, number)
    print(random_number)
    try:
        while guessLimit > 0:
            guess = int(input("what's ur guess? "))
            guessLimit -= 1
            if random_number == guess:
                print("you got right :)")
                break
            elif guess > random_number:
                print("you got wrong :|")
                print(f' {guessLimit} guess(es) left')
            else:
                print("you lost :(")
                print(f' {guessLimit} guess(es) right')
        print("game over!!")
        print(f'The no was {random_number} ')
    except ValueError:
                print("integers only!!")



def easy():
    print("choose between 1 - 10, you have 6 guesses")
    guessingGame(6,10)


def medium():
    print("choose between 1 - 20, you have 4 guesses")
    guessingGame(4,20)


def hard():
    print("choose between 1 - 50, you have 6 guesses")
    guessingGame(3,50)


def try_again():
    again = input("do you want to play again? Yes/No ")
    if again.upper() == "YES":
        welcome()
    elif again.upper() == "NO":
        print("bye bye")
    else:
        try_again()
         

def welcome():
    print("welcome")
    difficulty = input("choose difficulty level between easy, medium and hard")
    if difficulty.upper() == "EASY":
        easy()
        try_again()
    elif difficulty.upper() == "MEDIUM":
        medium()
        try_again()
    elif difficulty.upper() == "HARD":
        hard()
        try_again()

welcome()

```
